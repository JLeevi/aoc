```python
input_file = "data/input.txt"

table1 = {
    "A": 14,
    "K": 13,
    "Q": 12,
    "J": 11,
    "T": 10,
}

table2 = {
    "A": 14,
    "K": 13,
    "Q": 12,
    "T": 10,
    "J": 1,
}

def convert(card, table):
    if card.isdigit():
        return int(card)
    return table[card]

def get_strength1(cards):
    counts = []
    for card in set(cards):
        n_same = sum([1 for c in cards if c == card])
        counts.append(n_same)
    if 5 in counts:
        return 6
    if 4 in counts:
        return 5
    if 3 in counts and 2 in counts:
        return 4
    if 3 in counts:
        return 3
    if sum(1 for count in counts if count == 2) == 2:
        return 2
    if 2 in counts:
        return 1
    return 0

def get_strength2(cards):
    n_jokers = sum([1 for c in cards if c == 1])
    counts = []
    for card in set(cards):
        if card > 1:
            n_same = sum([1 for c in cards if c == card])
            counts.append(n_same)

    max_count = max([*counts, 1])
    max_count = max_count + n_jokers
    if max_count >= 5:
        return 6
    if max_count >= 4:
        return 5

    two_pair = sum(1 for count in counts if count == 2) == 2
    if (3 in counts and 2 in counts) or (two_pair and n_jokers >= 1):
        return 4
    if (3 in counts) or (2 in counts and n_jokers >= 1) or (n_jokers >= 2):
        return 3
    if two_pair:
        return 2
    if 2 in counts or n_jokers == 1:
        return 1
    return 0
    

with open(input_file, 'r') as f:
    lines = f.readlines()
    hands = [l.split() for l in lines]
    hands1 = []
    hands2 = []
    for hand, bid in hands:
        cards1 = [convert(card, table1) for card in hand]
        cards2 = [convert(card, table2) for card in hand]
        strength1 = get_strength1(cards1)
        strength2 = get_strength2(cards2)
        hands1.append((strength1, cards1, int(bid)))
        hands2.append((strength2, cards2, int(bid)))

    hands1 = sorted(hands1)
    hands2 = sorted(hands2)

    ans1 = 0
    ans2 = 0
    for rank in range(1, len(hands1) + 1):
        _, _, bid1 = hands1[rank - 1]
        _, _, bid2 = hands2[rank - 1]
        ans1 += rank * bid1
        ans2 += rank * bid2
    print(f"{ans1 = }")
    print(f"{ans2 = }")
```
